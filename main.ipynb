```python
df.isna().sum()
df = df.fillna(0)
```

```python
gender_map = {
    'M': 0,
    'F': 1
}
outcome_map = {
    False: 0,
    True: 1
}
df['gender'] = df['gender'].map(gender_map)
df['outcome'] = df['outcome'].map(outcome_map)
```

```python
import math
def plot_distribution(dataset,cols=5, width=20, height=15, hspace=0.5, wspace=0.5):
    plt.style.use('seaborn-whitegrid')  
    fig = plt.figure(figsize=(width,height))
    fig.subplots_adjust(left=None,bottom=None,right=None,top=None,wspace=wspace,hspace=hspace) 
    rows = math.ceil(float(dataset.shape[1]) / cols)
    
    for i,column in enumerate(dataset.columns):  
        ax = fig.add_subplot(rows,cols,i+1)  
        # ax.set_title(column)#

        if dataset.dtypes[column] == np.object:  
            g = sns.countplot(y=column,data=dataset)  
            plt.xticks(rotation=25)
        else:
            g = sns.distplot(dataset[column]) 
            plt.xticks(rotation=25)
```

```python
import warnings
warnings.filterwarnings('ignore') 
plot_distribution(df, cols=3, width=20, height=20, hspace=0.6, wspace=0.5)
```

## 进行特征选择

```python
X = df.drop('outcome', axis=1) 
y = df['outcome']
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2

k = 59
selector = SelectKBest(chi2, k=k)
X_new = selector.fit_transform(abs(X), y)

scores = selector.scores_
p_values = selector.pvalues_

selected_features = []
alpha = 0.01

for i in range(len(p_values)):
    if p_values[i] < alpha:
        selected_features.append(i)

selected_X = X.iloc[:, selected_features]
selected_X
```

```python
selected_X.columns
```

```python

from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression


model = LinearRegression()

rfe = RFE(model, 25)

X_selected = rfe.fit_transform(selected_X, y)

selected_feats = rfe.support_
print("选择的特征：", selected_X.columns[selected_feats])

removed_feats = ~rfe.support_
print("被剔除的特征：", selected_X.columns[removed_feats])
```

```python
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score

svc = SVC()
lr = LogisticRegression()
dt = DecisionTreeClassifier()

svc_scores = cross_val_score(svc, X_selected, y, cv=5)
lr_scores = cross_val_score(lr, X_selected, y, cv=5)
dt_scores = cross_val_score(dt, X_selected, y, cv=5)

print("SVM Scores:", svc_scores.mean())
print("Logistic Regression Scores:", lr_scores.mean())
print("Decision Tree Scores:", dt_scores.mean())
```
